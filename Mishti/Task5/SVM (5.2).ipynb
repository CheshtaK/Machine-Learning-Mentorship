```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm

data = pd.read_csv('../Downloads/adult.csv')

data.head()
```

```python
Income = pd.get_dummies(data['income'],drop_first=True)
Income.head()
```

```python
data = pd.concat([data, Income, Sex],axis=1)
data
```

```python
data = data.drop('income',axis=1)
data = data.drop('workclass',axis=1)
data = data.drop('education',axis=1)
data = data.drop('marital.status',axis=1)
data = data.drop('occupation',axis=1)
data = data.drop('relationship',axis=1)
data = data.drop('race',axis=1)
data = data.drop('native.country',axis=1)
data = data.drop('sex',axis=1)
data
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm

X = data.drop('>50K', axis=1)
y = data['>50K']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=24)
```

```python
clf = svm.SVC()

clf.fit(X_train,y_train)
```

```python
preds = clf.predict(X_test)
clf.score(X_test, y_test)
```

```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm

data = pd.read_csv('../Downloads/adult.csv')

data.head()
```

```python
Income = pd.get_dummies(data['income'],drop_first=True)
Income.head()
```

```python
Sex = pd.get_dummies(data['sex'],drop_first=True)
Sex.head()
```

```python
WClass = pd.get_dummies(data['workclass'],drop_first=True)
WClass.head()
```

```python
Edu = pd.get_dummies(data['education'])
Edu.head()
```

```python
ms = pd.get_dummies(data['marital.status'])
ms.head()
```

```python
occ = pd.get_dummies(data['occupation'],drop_first=True)
occ.head()
```

```python
rs = pd.get_dummies(data['relationship'],drop_first=True)
rs.head()
```

```python
Race = pd.get_dummies(data['race'],drop_first=True)
Race.head()
```

```python
nc = pd.get_dummies(data['native.country'],drop_first=True)
nc.head()
```

```python
data = pd.concat([data, Income, Sex, ms, occ, rs, nc, WClass, Edu, Race],axis=1)
data
```

```python
data = data.drop('income',axis=1)
data = data.drop('workclass',axis=1)
data = data.drop('education',axis=1)
data = data.drop('marital.status',axis=1)
data = data.drop('occupation',axis=1)
data = data.drop('relationship',axis=1)
data = data.drop('race',axis=1)
data = data.drop('native.country',axis=1)
data = data.drop('sex',axis=1)
data.head()
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm

X = data.drop('>50K', axis=1)
y = data['>50K']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=24)

clf = svm.SVC()

clf.fit(X_train,y_train)
```

```python
preds = clf.predict(X_test)
clf.score(X_test, y_test)
```
