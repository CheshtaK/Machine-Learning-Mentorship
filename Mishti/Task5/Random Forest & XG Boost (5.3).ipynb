```python
import matplotlib.pyplot as plt
```

```python
def plot_corr(df,size=10):


    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns);
    plt.yticks(range(len(corr.columns)), corr.columns);
```

```python
x=plot_corr(df)
```

```python
df.drop('workclass_?',axis=1,inplace=True)
df.drop('native.country_Scotland',axis=1,inplace=True)
df.drop('income_<=50K',axis=1,inplace=True)
```

```python
df.columns
```

```python
df.drop('marital.status_Widowed',axis=1,inplace=True)
df.drop('education_HS-grad',axis=1,inplace=True)
df.drop('occupation_Exec-managerial',axis=1,inplace=True)
```

```python
df.drop('relationship_Not-in-family',axis=1,inplace=True)
df.drop('race_White',axis=1,inplace=True)
df.drop('sex_Female',axis=1,inplace=True)
```

```python
df.head()
```

```python
X_int=df.drop('income_>50K',axis=1).values
Y_int=df['income_>50K'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_int,Y_int, test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=1000,max_depth=2,random_state=13)
clf.fit(X_train,y_train)
clf.predict(X_test)
scr=clf.score(X_test,y_test)
print(scr)
```

```python
import xgboost as xgb

xg = xgb.XGBClassifier(objective='binary:logistic', n_estimators=500,seed=1)
xg.fit(X_train,y_train)
predict=xg.predict(X_test)
print(xg.score(X_test,y_test))
```
