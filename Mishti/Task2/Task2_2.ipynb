```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt

stock=pd.read_csv('C:\\Users\\abc\\Documents\\GitHub\\STOCKDATA\\Machine-Learning-Mentorship\\Dataset\\stocks.csv', index_col=0)
df = pd.DataFrame(stock)
print(df.head())
```

```python
plotting = df.reindex(columns= ['MSFT','GOOG','SBUX','ADBE','NFLX','ORCL'])
print(plotting.head())
```

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
plotting['MSFT'].plot(kind = 'line',color='m')
plt.title('MSFT')

plt.subplot(132)
plotting['GOOG'].plot(kind = 'line',color='r')
plt.title('GOOG')

plt.subplot(133)
plotting['SBUX'].plot(kind = 'line',color='b')
plt.title('SBUX')

plt.show()

plt.figure(figsize=(15,5))
plt.subplot(131)
plotting['ADBE'].plot(kind = 'line',color='c')
plt.title('ADBE')

plt.subplot(132)
plotting['NFLX'].plot(kind = 'line',color='g')
plt.title('NFLX')

plt.subplot(133)
plotting['ORCL'].plot(kind = 'line',color='y')
plt.title('ORCL')

plt.show()
```
