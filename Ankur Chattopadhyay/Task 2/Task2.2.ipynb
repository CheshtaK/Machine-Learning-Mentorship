```python
data = pd.read_csv('../../Dataset/stocks.csv')
data.shape
```

```python
data.head()
```

*Changing dates to a better format*

```python
data.Date.dtype
```

```python
from datetime import datetime

def date_format(t):
    return (datetime.strptime(t, "%d-%b-%y").date())
```

```python
data.Date = data.Date.apply(date_format)
# data.Date
```

```python
data.head()
```

*Creating colors list to chose from*

```python
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']
```

```python
len(colors)
```

```python
plt.figure(figsize=(20,20),facecolor='w')
for i in range(1,len(data.columns)):
    col = data.columns[i]
    plt.subplot(4,3,i)
    plt.plot('Date', col, data=data, color=colors[i%10])
    plt.title(col)
plt.suptitle('Stock price comparision 2007-2017',fontsize=30)
# plt.show()
plt.savefig('stocks.png')
```
