```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
dataset_url = 'train.csv'
data1 = pd.read_csv(dataset_url)
```

```python
dataset_url = 'test.csv'
data2 = pd.read_csv(dataset_url)
```

```python
data1
```

```python
X_train=data1[['feature_1','feature_2','feature_3','feature_4','feature_5']]
Y_train=data1['target']
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train,Y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
data2
```

```python
X_test=data2[['feature_1','feature_2','feature_3','feature_4','feature_5']]
Y_test=data2['target']
```

```python
pred = reg.predict(X_test)
```

```python
def r2(pred,y):
 sst = np.sum((y-y.mean())**2)
 ssr = np.sum((pred-y)**2)
 r2 = 1-(ssr/sst)
 return(r2)
#----------------
r2(pred,Y_test)
```
