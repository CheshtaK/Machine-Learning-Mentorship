```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
data = pd.read_csv('adult.data.csv',names=columns,sep=',\s',na_values=["?"])
data1 = pd.read_csv('adult.test.csv',names=columns,sep=',\s',na_values=["?"])
```

```python
data['capital-gain'] = data['capital-gain'].replace(0,np.nan)
data1['capital-gain'] = data1['capital-gain'].replace(0,np.nan)
```

```python
data['capital-loss'] = data['capital-loss'].replace(0,np.nan)
data1['capital-loss'] = data1['capital-loss'].replace(0,np.nan)
```

```python
#data['native-country'] = data['native-country'].replace('?',np.nan)
```

```python
data.head(30)
```

```python
mean_value_gain=data['capital-gain'].mean()
data['capital-gain']=data['capital-gain'].fillna(mean_value_gain)

mean_value_loss=data['capital-loss'].mean()
data['capital-loss']=data['capital-loss'].fillna(mean_value_loss)

mean_value_gain=data1['capital-gain'].mean()
data1['capital-gain']=data1['capital-gain'].fillna(mean_value_gain)

mean_value_loss=data1['capital-loss'].mean()
data1['capital-loss']=data1['capital-loss'].fillna(mean_value_loss)

```

```python
data.head()
```

```python
data = data.drop(['marital-status','relationship', 'race','native-country'], axis=1)
data1 = data1.drop(['marital-status','relationship', 'race','native-country'], axis=1)
```

```python
labelEncoder = LabelEncoder()
labelEncoder.fit(data['sex'])
labelEncoder.fit(data1['sex'])
data['sex'] = labelEncoder.transform(data['sex'])
data1['sex'] = labelEncoder.transform(data1['sex'])
```

```python
data
```

```python
X_train= data.drop('income',axis=1)
X_test= data['income']
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
