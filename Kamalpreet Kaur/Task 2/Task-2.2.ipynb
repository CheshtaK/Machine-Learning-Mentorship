```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset_url= 'stocks.csv'
data3 = pd.read_csv(dataset_url)
data3
```

```python
data3['Year'] = pd.DatetimeIndex(data3['Date']).year
data3['Year']
```

```python
fig, ax=plt.subplots(nrows=2,ncols=3 , figsize=(15,8))

ax[0,0].plot(data3['Date'],data3['MSFT'],'g')
ax[0,0].set_title("MSFT")

ax[0,1].plot(data3['Date'],data3['GOOG'],'indigo')
ax[0,1].set_title("GOOG")

ax[0,2].plot(data3['Date'],data3['SBUX'],'fuchsia')
ax[0,2].set_title("SBUX")

ax[1,0].plot(data3['Date'],data3['ADBE'], 'gold')
ax[1,0].set_title("ADBE")

ax[1,1].plot(data3['Date'],data3['NFLX'], 'goldenrod')
ax[1,1].set_title("NFLX")

ax[1,2].plot(data3['Date'],data3['ORCL'], 'teal')
ax[1,2].set_title("ORCL")

plt.show()
```
